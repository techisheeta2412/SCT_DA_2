## Dataset Inspection

The dataset is inspected to understand its dimensions, column structure, and data types before performing data cleaning.

```python
print("Dataset Shape:", df.shape)
print("Number of Rows:", df.shape[0])
print("Number of Columns:", df.shape[1])
```

## Missing Value Identification

Missing values are identified to detect incomplete records in the dataset before performing data cleaning.

```python
missing_values = df.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)

print("\nColumns with Missing Values:")
print(missing_values[missing_values > 0])
```

```python
df[df["Postal Code"].isnull()]
```

```python
df[df["Postal Code"].isnull()][["City", "State", "Postal Code"]]
```

## Handling Missing Values

The Postal Code column contained 11 missing values. The corresponding City and State values were inspected before handling the missing data. The missing records belonged to Burlington, Vermont, so the appropriate postal code was used to fill the missing values.

```python
df["Postal Code"] = df["Postal Code"].fillna(5401)
```

```python
print("Missing Values After Cleaning:")
print(df.isnull().sum())

print("\nTotal Missing Values:", df.isnull().sum().sum())
```

## Duplicate Detection and Removal

Duplicate records are identified and removed to improve data quality and prevent repeated observations from affecting the analysis.

```python
duplicate_count = df.duplicated().sum()

print("Number of Duplicate Rows:", duplicate_count)
```

```python
df = df.drop_duplicates()

print("Duplicate Rows After Cleaning:", df.duplicated().sum())
print("Dataset Shape After Removing Duplicates:", df.shape)
```

## Data Type Conversion

The Order Date and Ship Date columns were stored as object data types. These columns are converted to datetime format to enable accurate date-based analysis.

```python
print("Data Types Before Conversion:")

print(df[["Order Date", "Ship Date"]].dtypes)
```

```python
df["Order Date"] = pd.to_datetime(
    df["Order Date"],
    format="%d/%m/%Y"
)

df["Ship Date"] = pd.to_datetime(
    df["Ship Date"],
    format="%d/%m/%Y"
)

print("Date columns converted successfully.")
```

```python
print("Data Types After Conversion:")

print(df[["Order Date", "Ship Date"]].dtypes)
```

```python
df.head()
```

## Duplicate Row Handling

Duplicate rows were identified and removed to improve data quality and prevent repeated records in the cleaned dataset.

```python
print("Duplicate Rows Before Cleaning:", df.duplicated().sum())

df = df.drop_duplicates()

print("Duplicate Rows After Cleaning:", df.duplicated().sum())
```

```python
print("FINAL DATA CLEANING SUMMARY")

print("\nDataset Shape:")
print(df.shape)

print("\nTotal Missing Values:")
print(df.isnull().sum().sum())

print("\nTotal Duplicate Rows:")
print(df.duplicated().sum())

print("\nData Types:")
print(df.dtypes)
```

## Export Cleaned Dataset

The cleaned and prepared dataset is exported to a new CSV file for further data analysis and visualization.

```python
df.to_csv("cleaned_superstore.csv", index=False)

print("Cleaned dataset exported successfully!")
```
